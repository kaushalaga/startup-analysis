```python
new_data=data.groupby('')['AmountInUSD'].count().sort_values(ascending=False)
```

```python
act_amount=[]
for amount in data['AmountInUSD']:
    try:
        act_amount.append(round(int(amount.replace(",",""))))
    except:
        act_amount.append(amount)
data['AmountInUSD']= act_amount
```

```python
data['AmountInUSD'].value_counts().plot(kind='bar',legend=True)
```
